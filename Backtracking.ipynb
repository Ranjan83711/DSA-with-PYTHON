```python
# Generate Parenthesis
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        def solve(index,total,brackets,result):
            if index>=len(brackets):
                if total==0:
                    result.append("".join(brackets))
                    return
                if total>len(brackets)//2:
                    return
                elif total<0:
                    return
                brackets[index]='('
                Sum=total+1
                solve(index+1,Sum)
                brackets[index]=')'
                Sum=total-1
                solve(index+1,Sum)
        solve(0,0,0,0)
        return result
```

```python
# Combination Sum I
class Solution:
    def combinationSum(self, candidates, target):
# candidates: numbers hum use kar sakte hain
# target: jis sum ko achieve karna hai
        result = []
        
        def backtrack(i, curr, total):
            # If total == target: valid combination
            if total == target:
                result.append(curr.copy())
                return
            
            # If out of bounds or total exceeds target: stop
            if i >= len(candidates) or total > target:
                return
            
            # 1Ô∏è‚É£ Pick candidates[i]
            curr.append(candidates[i])
            backtrack(i, curr, total + candidates[i])
            curr.pop()     # backtrack
            
            # 2Ô∏è‚É£ Skip candidates[i]
            backtrack(i + 1, curr, total)
        
        backtrack(0, [], 0)
        return result
```
