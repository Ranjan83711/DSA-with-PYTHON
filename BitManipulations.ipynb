```python
# Min bit flips to convert Number 
start=10
end=7
result=start^end
print(result.bit_count()) 

# We do NOT check whether it is 0→1 or 1→0.

# We only check if the bits are different.

# XOR marks all different bits with 1.

# So counting 1s in XOR tells exactly how many flips happened

# we are counting 1 only because in XOR operator diff bits are 1 and same bits are 0
```

```python
# Subsets of number
nums=[1,2,3]
n=len(nums)
total_subsets=1<<n
result=[]
for num in range(0,total_subsets):
    lst=[]
    for i in range(0,n):
        if num & (1<<i)!=0:
            lst.append(nums[i])
    result.append(lst)
print(result)
```
