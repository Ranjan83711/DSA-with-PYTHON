```python
# 7. Reverse Integer
class Solution:
    def reverse(self, x: int) -> int:
        count=0
        while x!=0:
            ld=x%10 if x > 0 else int(x % -10)
            count=(count*10)+ld
            x=int(x/10)
        if count < -2**31 or count > 2**31:
            return 0
        return count
```

```python
9. Palindrome Number
class Solution:
    def isPalindrome(self, x: int) -> bool:
        if x<0:
            return False
        count=0
        original=x
        while x>0:
            ld=x%10 
            count=(count*10)+ld
            x=int(x/10)
        return count==original
```

```python
# Optimal method 

def threeSum(self, nums: List[int]) -> List[List[int]]:
    nums.sort()          # Step 0: Sort the list in ascending order
    n = len(nums)        # Step 1: Find the length of the list
    ans = []             # Step 2: This will store all the triplets

    for i in range(n-2):  # Step 3: Loop through each number as the first number of the triplet
        if i > 0 and nums[i] == nums[i-1]:  # Step 4: Skip duplicates for the first number
            continue

        j = i + 1        # Step 5: Start pointer j just after i
        k = n - 1        # Step 6: Start pointer k at the end of the list

        while j < k:     # Step 7: Loop while the two pointers do not cross each other
            total_sum = nums[i] + nums[j] + nums[k]  # Step 8: Add three numbers

            if total_sum < 0:   # Step 9: If sum is too small, move j forward to increase sum
                j += 1
            elif total_sum > 0: # Step 10: If sum is too big, move k backward to decrease sum
                k -= 1
            else:                # Step 11: If sum is exactly zero, we found a triplet
                temp = [nums[i], nums[j], nums[k]]  # Step 12: Store it in a temporary list
                ans.append(temp)                    # Step 13: Add this triplet to the answer

                j += 1  # Step 14: Move j forward to check next possibility
                k -= 1  # Step 15: Move k backward to check next possibility

                # Step 16: Skip duplicates for j
                while j < k and nums[j] == nums[j-1]:
                    j += 1
                # Step 17: Skip duplicates for k
                while j < k and nums[k] == nums[k+1]:
                    k -= 1

    return ans  # Step 18: Return the list of all unique triplets
```

```python
# --- Input list ---
nums = [-1, 0, 1, 2, -1, -4]   # You can change this to test other lists

# Step 0: Sort the list in ascending order
nums.sort()
print("Sorted List:", nums)

# Step 1: Find the length of the list
n = len(nums)

# Step 2: This will store all the triplets
ans = []

# Step 3: Loop through each number as the first number of the triplet
for i in range(n - 2):
    # Step 4: Skip duplicates for the first number
    if i > 0 and nums[i] == nums[i - 1]:
        continue

    # Step 5: Start pointer j just after i
    j = i + 1

    # Step 6: Start pointer k at the end of the list
    k = n - 1

    # Step 7: Loop while the two pointers do not cross each other
    while j < k:
        # Step 8: Add three numbers
        total_sum = nums[i] + nums[j] + nums[k]

        # Step 9: If sum is too small, move j forward to increase sum
        if total_sum < 0:
            j += 1
        # Step 10: If sum is too big, move k backward to decrease sum
        elif total_sum > 0:
            k -= 1
        else:
            # Step 11: If sum is exactly zero, we found a triplet
            temp = [nums[i], nums[j], nums[k]]   # Step 12
            ans.append(temp)                     # Step 13

            # Step 14 & 15: Move both pointers
            j += 1
            k -= 1

            # Step 16: Skip duplicates for j
            while j < k and nums[j] == nums[j - 1]:
                j += 1

            # Step 17: Skip duplicates for k
            while j < k and nums[k] == nums[k + 1]:
                k -= 1

# Step 18: Print the list of all unique triplets
print("Unique triplets that sum to 0 are:")
print(ans)
```

```python
# 4Sum problem  TC=O(n****4)

# Bruteforce method

nums=[1,0,-1,0,-2,2,5,9]
n=len(nums)
my_set=set()
for i in range(0,n):
    for j in range(i+1,n):
        for k in range(j+1,n):
            for l in range(k+1,n):
                total=nums[i]+nums[j]+nums[k]+nums[l]
                if total==0:
                    temp=[nums[i],nums[j],nums[k],nums[l]]
                    temp.sort()
                    my_set.add(tuple(temp))

result=[]
for ans in my_set:
    result.append(list(ans))
print(result)

```

```python
nums = [1, 0, -1, 0, -2, 2]
target = 0
nums.sort()
n = len(nums)
ans = []

for i in range(n - 3):
    if i > 0 and nums[i] == nums[i - 1]:
        continue
    for j in range(i + 1, n - 2):
        if j > i + 1 and nums[j] == nums[j - 1]:
            continue

        k, l = j + 1, n - 1

        while k < l:
            total = nums[i] + nums[j] + nums[k] + nums[l]

            if total == target:
                ans.append([nums[i], nums[j], nums[k], nums[l]])
                k += 1
                l -= 1

                while k < l and nums[k] == nums[k - 1]:
                    k += 1
                while k < l and nums[l] == nums[l + 1]:
                    l -= 1

            elif total < target:
                k += 1
            else:
                l -= 1

print(ans)
```

```python
# Intro to binary search
#Iterative method
nums=[2,3,6,4,41,10,54,19]
target=19
def binarySearch(nums,target):
    nums.sort()
    n=len(nums)
    low=0
    high=n-1
    while low<=high:
        mid=(high+low)//2
        if nums[mid]==target:
            return mid
        elif nums[mid]<target:
            low=mid+1
        else:
            high=mid-1
    return -1

binarySearch(nums,target)
```

```python
#Recursive method

nums=[2,3,6,4,19,41,10,54]
target=19
low=0
high=len(nums)-1

def recursivebinarySearch(nums,low,high):
    if low>high:
        return -1
    mid=(low+high)//2
    if nums[mid]==target:
        return mid
    elif nums[mid]<target:
        return recursivebinarySearch(nums,mid+1,high)
    else:
        return recursivebinarySearch(nums,low,mid-1)
a=recursivebinarySearch(nums,low,high)
print(a)
```
