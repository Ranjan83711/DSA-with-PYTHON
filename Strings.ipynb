```python
# Remove Outermost parenthesis

s='(()())(())'
result=''
count=0
for i in s:
    if i=='(':
        count+=1
        if count>1:
            result+=i
    else:
        count-=1
        if count>0:
            result+=i
print(result)
```

```python
str1=['flower','flow','flight']
def fun(str1):
    result=''
    base=str1[0]
    for i in range(0,len(base)):
        for ch in str1[1:]:
            if i==len(ch) or ch[i]!=base[i]:
                return result
        result+=base[i]
fun(str1)       
```

```python
# isomorphic strings
s = "paper"
t = "title"
mapping_st={}
mapping_ts={}
def isomorphic(s,t):
    if len(s)!=len(t):
        return False
    for i in range(len(s)):
        char_s=s[i]
        char_t=t[i]
        if char_s in mapping_st:
            if mapping_st[char_s]!=char_t:
                return False
        else:
            mapping_st[char_s]=char_t
        if char_t in mapping_ts:
            if mapping_ts[char_t]!=char_s:
                return False
        else:
            mapping_ts[char_t]=char_s
    return True
isomorphic(s,t)
```

```python
# Rotate string
s= 'abcde'
goal='cdeab'
def rotate(s,goal):
    for i in s:
        s+=i
        if goal in s:
            return True
    return False
rotate(s,goal)
```

```python
#Sort characters by string 
s='Aabb'
result=''
maped={}
for i in s:
    maped[i]=maped.get(i,0)+1 
sorted_map=sorted(maped.items(),key=lambda x:(-x[1],x[0]))
sorted_map
for ch,freq in sorted_map:
    result=result+(ch*freq)
result
```

```python
# Max Nesting depth of a Parenthesis
s = "(1+(2*3)+((8)/4))+1"
count=0
for i in s:
    for j in s:
        if i=='(':
            if j==')':
                count+=1
                break
count
```

```python
# Max Nesting depth of a Parenthesis

count=0
max_count=0
for ch in s:
    if ch=='(':
        count+=1
        max_count=max(count,max_count)
    elif ch==')':
        count-=1
max_count
```

```python
# Roman to integer 
letter={'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}
s='MCMXCIV'
n=len(s)
result=0
for i in range(0,n):
    if i<n-1 and letter[s[i]]<letter[s[i+1]]:
        result-=letter[s[i]]
    else:
        result+=letter[s[i]]
result
```

```python
def romanToInt(s):
    roman = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}
    total = 0
    prev = 0
    for ch in reversed(s):
        if roman[ch] < prev:
            total -= roman[ch]
        else:
            total += roman[ch]
        prev = roman[ch]
    return total
romanToInt(s)
```
