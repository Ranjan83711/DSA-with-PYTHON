```python
# Next Permutation
nums=[2,3,1]
n = len(nums)

# Step 1: find pivot
i = n - 2
while i >= 0 and nums[i] >= nums[i + 1]:
    i -= 1

# Step 2: find successor
if i >= 0:
    j = n - 1
    while nums[j] <= nums[i]:
        j -= 1
    nums[i], nums[j] = nums[j], nums[i]

# Step 3: reverse suffix
nums[i + 1:] = reversed(nums[i + 1:])
nums
```

```python
# Merge Overlapping Subintervals
intervals =[[4,7],[1,4]]
def merge(intervals):
    if not intervals:
        return []

    # Step 1: sort
    intervals.sort(key=lambda x: x[0])

    merged = []

    # Step 2: single pass
    for interval in intervals:
        if not merged or interval[0] > merged[-1][1]:
            merged.append(interval)
        else:
            merged[-1][1] = max(merged[-1][1], interval[1])

    return merged
merge(intervals)
```

```python
# Floor and Ceil in Sorted Array
nums =[3, 4, 4, 7, 8, 10]
x=5
def floor_ceil(nums,x):
    low=0
    high=len(nums)-1
    floor,ceil=0,0
    for i in range(len(nums)):
        mid=(low+high)//2
        if nums[mid]==x:
            return [mid,mid]
        elif nums[mid]<x:
            floor=nums[mid]
            low=mid+1
        else:
            ceil=nums[mid]
            high=mid-1
    return floor,ceil
floor_ceil(nums,x)                  
```

```python
# First and last occurrence
nums = [5, 7, 7, 8, 8, 10]
target=8
def final(nums,target):
    def first(nums,target):
        low=0
        high=len(nums)-1
        first=-1
        for i in range(len(nums)):
            mid=(low+high)//2
            if nums[mid]==target:
                first=mid
                high=mid-1
            elif nums[mid]<target:
                low=mid+1
            else:
                high=mid-1
        return first
    
    def last(nums,target):
        low=0
        high=len(nums)-1
        last=-1
        for i in range(len(nums)):
            mid=(low+high)//2
            if nums[mid]==target:
                first=mid
                low=mid+1
            elif nums[mid]<target:
                low=mid+1
            else:
                high=mid-1
        return first
    return first(nums, target), last(nums, target)
final(nums,target)
```

```python
# Search in rotated sorted array-I
nums = [4, 5, 6, 7, 0, 1, 2]
target=0
def search(nums,target):
    low=0
    high=len(nums)-1
    while low<=high:
        mid=(low+high)//2
        if nums[mid]==target:
            return mid
        if nums[low]<=nums[mid]:
            if nums[low]<=target<=nums[mid]:
                high=mid-1
            else:
                low=mid+1
        else:
            if nums[mid]<target<=nums[high]:
                low=mid+1
            else:
                high=mid-1
    return -1
search(nums,target)
```
